Una empresa desea analizar sus ventas mensuales durante un año para entender mejor la
tendencia de ventas y tomar decisiones informadas. Se proporciona un archivo CSV llamado
"ventas_mensuales.csv" que contiene dos columnas: "Mes" y "Ventas". Sin embargo, el
archivo CSV tiene datos incompletos en algunas entradas. Se te solicita:

* Leer los datos del archivo CSV y explorar la estructura y contenido de los mismos.
* Manejar los datos incompletos: puedes elegir entre eliminar los registros con datos faltantes o rellenar los valores faltantes utilizando alguna técnica común en análisis de datos, como, por ejemplo, rellenar con la media de las ventas mensuales.
* Calcular la media y la desviación estándar de las ventas mensuales actualizadas.
* Graficar las ventas mensuales actualizadas utilizando matplotlib.

# Paso 1: Importa las bibliotecas necesarias y lee los datos del archivo CSV.

```python
# Paso 1: Importa las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Lee los datos del archivo CSV

```python
datos_ventas = pd.read_csv("ventas_mensuales.csv")
```

# Paso 2: Explora los datos para entender su estructura y contenido.

```python
# Paso 2: Explora los datos
print(datos_ventas.head()) # Muestra las primeras filas de los datos
print(datos_ventas.info()) # Muestra información sobre los datos
```

# Paso 3: Maneja los datos incompletos. 

En este caso, vamos a rellenar los valores faltantes con la media de las ventas mensuales.

```python
# Paso 3: Maneja los datos incompletos
datos_ventas['Ventas'].fillna(datos_ventas['Ventas'].mean(), inplace=True)
```

# Paso 4: Calcula la media y la desviación estándar de las ventas mensuales actualizadas.

```python
# Paso 4: Calcula la media y la desviación estándar de las ventas mensuales actualizadas
media_ventas = np.mean(datos_ventas['Ventas'])
desviacion_estandar_ventas = np.std(datos_ventas['Ventas'])
print("La media de las ventas mensuales es:", media_ventas)
print("La desviación estándar de las ventas mensuales es:", desviacion_estandar_ventas)
```

# Paso 5: Grafica las ventas mensuales actualizadas.

```python
# Paso 5: Grafica las ventas mensuales actualizadas
plt.figure(figsize=(10, 6))
plt.plot(datos_ventas['Mes'], datos_ventas['Ventas'], marker='o',
color='b', linestyle='-')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
